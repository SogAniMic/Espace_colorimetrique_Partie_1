## Espace HSV ou TSV 

HSV : Hue - Saturation - Value 

TSV : Teinte - Saturation - Valeur

### La teinte 

La teinte est codée suivant l'angle qui lui correspond sur le cercle de couleur 

- 0 ou 360 -> Rouge 
- 60       -> Jaune
- 120      -> Vert
- 180      -> Cyan
- 240      -> Bleu
- 300      -> Magenta



### La saturation

La saturation désigne l'intensité de la couleur 

- Varie entre 0 et 100 % 
- Plus la saturation d'une couleur est faible, plus l'image sera grise 


## La valeur 

La valeur est la brillance de la couleur 

- Varie entre 0 et 100 %
- Plus la valeur d'une couleur est faible, plus la couleur est sombre 
- Une valeur de 0 correspond au noir 

```python
# Exemple 3 : Nous convertissons les pixels de l'image du papillon au format HSV
```

```python
from skimage.color import rgb2hsv
```

```python
hsv_papillon = rgb2hsv(rgb_papillon)
```

```python
imshow (hsv_papillon, cmap = "hsv")
```

# Les images monocanales ou monochrone 

Pour une image en couleur au format RGB, u,e image monocanale est obtenue soit : 
- en conservant uniquement le canal rouge de l'image
- en conservant uniquement le canal vert de l'image
- en conservant uniquement le canal bleu de l'image
- en calculant une moyenne des 3 canaux RGB de sorte à obtenir une seule valeur (l'image obtenue ici est une image en nuance de gris )

Pour une image en couleur au format HSV, une image monocanale est obtenue soit :
- en conservant uniquement le canal teinte 
- en consercant uniquement le canal saturation 
- en conservant unqiuement le canal valeur 

```python
# Créer une copie du tableau de pixels rgb_papillon 
```

```python
rgb_papillon2 = rgb_papillon.copy()
```

# Image monocanale rouge 

```python
# Récupérer le canal rouge de chaque pixel
```

```python
rgb_papillon2_R = rgb_papillon2[:,:,0]
```

```python
# Afficher l'image monocanale rouge
```

```python
imshow(rgb_papillon2_R, cmap = "Reds")
```

# Image monocanale vert

```python
# Recupérer le canal vert de chaque pixel
```

```python
rgb_papillon2_G = rgb_papillon2[:,:,1]
```

```python
# Afficher l'image monocanale vert
imshow(rgb_papillon2_G, cmap = "Greens")
```

```python
# Recupérer le canal bleu de chaque pixel
rgb_papillon2_B = rgb_papillon2[:,:,2]
# Afficher l'image monocanale verte
imshow(rgb_papillon2_B, cmap = "Blues")
```

# Représentation complète

```python
fig,ax = plt.subplots(1,4,figsize = (20,4))
ax[0].imshow(rgb_papillon2)
ax[0].set_title('Original')
ax[1].imshow(rgb_papillon2_R, cmap = "Reds")
ax[1].set_title('Canal Rouge')
ax[2].imshow(rgb_papillon2_G, cmap = "Greens")
ax[2].set_title('Canal Vert')
ax[3].imshow(rgb_papillon2_B, cmap = "Blues")
ax[3].set_title('Canal Bleu')
```
